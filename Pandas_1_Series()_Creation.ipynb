#### **1. Pandas Series :** 
##### A Pandas Series is one-dimensional labeled array that is capable of holding data of any type (integer, string, float, Python objects etc.) of collection along with index and data type. 
##### A 1-dimensional labeled array,Series is nothing but the column in Excel

##### **Syntax :pd.Series(data, index, dtype)**
##### **Syntax : pd.Series(data=None, index=None, dtype=None, name=None, copy=False)**
##### **Parameters:**
##### **data** → values (list, tuple, dict, numpy array)
##### **index** → labels (optional)
##### **dtype** → data type (optional)**
##### **name** → name of the column (optional)
##### **copy=True** → It creates the copy of the original data/array/collection (optional)
##### What does copy mean? It decides whether: Pandas should use the original data or not

##### **Copy :** It is used to copy the original array or modifies the same array if Copy=False
| copy  | Meaning                       |
| ----- | ----------------------------- |
| False | Use same memory (if possible) |
| True  | Create new copy in memory     |

##### **Copy=True** : Means it will copies the Original data/values/collection/array,It won't affect if the original data/array will change
##### **Copy=False** : Means it will not copies the Original data/values/collection/array,It will directly affect the original data/array will change

```python
import pandas as pd
import numpy as np
arr = np.random.randint(50,80,5)
df = pd.Series(arr,index=np.arange(5),name='Marks',copy=True)
print(arr)
df.to_frame()
```

##### After modifying the original array or data it will not change in the Series()

```python
arr[0]=100
print(arr)
df.to_frame()
```

##### **Copy=False** : Means it will not copies the Original data/values/collection/array,It will directly affect the original data/array will change

```python
import pandas as pd
import numpy as np
arr = np.random.randint(50,80,5)
df = pd.Series(arr,index=np.arange(5),name='Marks',copy=False)
print(arr)
df.to_frame()
```

##### After modifying the original array or data it will change in the Series(), Becuase we are not copied the original array

```python
arr[0]=100
print(arr)
df.to_frame()
```

##### In Series() We Can't Perform Muliple Series time at a time , to perform Multiple Series we need to use concat, So we can pass multiple Series() 

```python
import pandas as pd
import numpy as np
s1 = pd.Series(np.random.randint(50,100,4),index=[1,2,3,4],name='Marks1')
s2 = pd.Series(np.random.randint(50,100,4),index=[1,2,3,4],name='Marks2')
s3 = pd.Series(np.random.randint(50,100,4),index=[1,2,3,4],name='Marks3')
s4 = pd.Series(np.random.randint(50,100,4),index=[1,2,3,4],name='Marks4')
s5 = pd.Series(np.random.randint(50,100,4),index=[1,2,3,4],name='Marks5')
df = pd.concat([s1,s2,s3,s4,s5], axis=1)
df
```

```python
df[1:3]
```

##### **NOTE :**
##### Core Idea (Understand This Clearly)
##### Series → 1D (only one column) If we want to do with multiple columns we can use concat()
##### DataFrame → 2D (rows + columns)
##### Why Series cannot support 2D? -> A Series is designed to represent a single column only.
