##### **1.3 Access by using loc[] : Location** 
##### **Syntax : variable_name.loc[row_index]**
##### **Syntax : variable_name.loc[start_row_index:end_row_index]**
##### **Syntax : variable_name.loc[start_row_index:end_row_index:step_index]**
##### We can access using location i.e, loc[] by passing row_index.
##### In Series() Only Single Row Indexing or Slicing will perform
##### **NOTE:** Start and Stop Index are inclusive

```python
import pandas as pd
import numpy as np
df = pd.Series(data=[10,20,30,40,50],index=['L','B','D','C','A'],dtype=int,name='Sales')
df.to_frame()
```

##### We can access or pass only one row throught indexing

```python
print(df.loc['B'])
```

##### If we trying to access the multiple rows through indexing it will raise an error

```python
print(df.loc['B','A'])
```

##### We can access multiple rows through slicing only
##### **Note :** Here Start Index and Stop Index Both are Inclusive

```python
print(df.loc['B':'A'])
```

```python
print(df.loc['B':'A':2])
```

##### If we want to access any row through index or slicing , make sure that , Whatever indexing we are passed, only that indexing values we can pass inside square brackets []

```python
import pandas as pd
import numpy as np
df = pd.Series(data=[10,20,30,40,50],index=[0,1,2,3,4],dtype=int,name='Sales')
df.to_frame()
```

```python
print(df.loc[0:4:2])
```

##### **1.4 Access by using iloc[] : Integer Location** 
##### **Syntax : variable_name.iloc[row_index]**
##### **Syntax : variable_name.iloc[start_row_index:end_row_index]**
##### **Syntax : variable_name.iloc[start_row_index:end_row_index:step_index]**
##### We can access using Integer location i.e, iloc[] by passing row_index.
##### In Series() Only Single Row Indexing or Slicing will perform
##### **NOTE:** Start is Inclusive and Stop Index is Exclusive in iloc[] and It only accepts the Integer Index or Slicing

```python
import pandas as pd
import numpy as np
df = pd.Series(data=[10,20,30,40,50],index=[0,1,2,3,4],dtype=int,name='Sales')
df.to_frame()
```

```python
print(df.iloc[0])
```

##### **Note :** Start_index is inclusive and Stop_index is Exclusive

```python
print(df.iloc[0:4])
```

```python
print(df.iloc[0:6:2])
```

##### Here also I can't pass multiple index values at a time

```python
print(df.iloc[0,3])
```

##### **1.6 Slicing**
##### **Syntax : variable_name[index]**
##### **Syntax : variable_name[start:stop]**
##### **Syntax : variable_name[start:stop:step]**
##### Slicing is used to get multiple values at a time and It returns the name and dtype
##### Slicing By Position : variable_name[1:2]
##### Slicing By Position : variable_name['A':'B']

```python
import pandas as pd
import numpy as np
df = pd.Series(data=[100,200,300,400,500],index=['A','B','C','D','E'],dtype=int,name='Marks')
df.to_frame()
```

```python
print(df[1])
print(df['B'])
```

```python
print(df[0:5])
```

```python
print(df['A':'E'])
```

```python
print(df['A':'E':2])
```

##### With the help of time module we can check which one will access faster than other But we will not get correct output

```python
import time 
start=time.time()
print(df.at['A'])
end=time.time()
print(f'Total Time {end-start}')
```

```python
import time 
start=time.time()
print(df.loc['A'])
end=time.time()
print(f'Total Time {end-start}')
```

##### With the help of timeit module we can check which one will access faster than other

```python
import timeit
stmt = "df.loc['A']"
setup = """
import pandas as pd
df = pd.Series({'A': 10, 'B': 20, 'C': 30})
"""
execution_time = timeit.timeit(stmt=stmt, setup=setup, number=100000)
print("Total Time:", execution_time)
```

```python
import timeit
stmt = "df.at['A']"
setup = """
import pandas as pd
df = pd.Series({'A': 10, 'B': 20, 'C': 30})
"""
execution_time = timeit.timeit(stmt=stmt, setup=setup, number=100000)
print("Total Time:", execution_time)
```

##### **2.1. Modify Single Value**
##### **Syntax : variable_name[index/position/label]=NewValue**

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name="Marks")
df.to_frame()
```

##### Here we can update values using index position.
##### **Note** : Suppose if the index is not present in the data/array/table, then by default it will add the new value to that index position

```python
df[0]=100
df.to_frame()
```

```python
df[5]=50
df.to_frame()
```

##### If we want to work with multiple columns we can merge we can do

```python
import pandas as pd
import numpy as np
df1 = pd.Series(data=np.random.randint(100,1001,10),index=np.arange(10),dtype=int,name='Growth')
df2 = pd.Series(data=np.random.randint(100,1001,10),index=np.arange(10),dtype=int,name='Growth')
df3= pd.Series(data=np.random.randint(100,1001,10),index=np.arange(10),dtype=int,name='Growth')
df4 = pd.Series(data=np.random.randint(100,1001,10),index=np.arange(10),dtype=int,name='Growth')
df5 = pd.Series(data=np.random.randint(100,1001,10),index=np.arange(10),dtype=int,name='Growth')
df=pd.concat([df1,df2,df3,df4,df5],axis=1)
df
```

##### **2.2. Modify Using loc[]**
##### **Syntax : variable_name.loc[index/position/label]=NewValue**

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name="Marks")
df.to_frame()
```

##### With the help of indexing we can modify single data/values

```python
df.loc[0]=100
df.to_frame()
```

##### With the help of slicing we can modify entire data/values

```python
df.loc[0:4]=100
df.to_frame()
```

##### If we want to add or insert the new row we can insert it directly

```python
df.loc[5]=200
df.to_frame()
```

##### **2.3. Modify Using iloc[]**
##### **Syntax : variable_name.loc[index(only integer)]=NewValue**

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name="Marks")
df.to_frame()
```

##### With the help of indexing we can modify single data/values

```python
df.iloc[0]=100
df.to_frame()
```

##### With the help of slicing we can modify entire data/values

```python
df.iloc[0:5]=100
df.to_frame()
```

##### If we want to add or insert the new row we can't insert in iloc[], iloc[] can't be changed their target

```python
df.iloc[5]=100
df.to_frame()
```

##### **2.4. Modify Multiple Values(Fancy Indexing)**

##### **2.5. Modify Using Boolean Indexing(Filtering)**
##### **Syntax : variable_name[condition]=NewValue**
##### **Note :** The NewValue will update on entire data

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name="Marks")
df.to_frame()
```

```python
df[df>70]=100
df.to_frame()
```

##### If we want to update single value we can use == double equal to so only that value will modifies

```python
df[df==61]=500
df.to_frame()
```

##### **2.8. Replace Values**

##### **2.9. Apply Function**
##### **Syntax : variable_name.apply(operation)**

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name="Marks")
df.to_frame()
```

```python
df=df.apply(lambda df: df + 10)
df.to_frame()
```

##### **2.10. Modify Data Type**
