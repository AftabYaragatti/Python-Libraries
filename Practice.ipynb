# 3.Array with Built-in Methods

```python
# Sort() : it sorts the array elements
import numpy as np
arr = np.random.randint(1,10,5)
print(arr)
print(np.sort(arr))
```

```python
# argsort() : it will returns the sorted indexs of array elements
arr = np.random.randint(1,10,5)
print(arr)
print(np.argsort(arr))
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.min(arr))
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.argmin(arr))
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.max(arr))
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.argmax(arr))
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.sum(arr))
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.average(arr))
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.mean(arr))
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.median(arr)) #Sorted array ka middle value , If even → average of two middle values
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.sqrt(arr)) 
```

```python
arr = np.random.randint(1,10,5) #Variance=n∑(x−mean)^2/total numbers
print(arr)
print(np.var(arr)) # mean=2.8 => (0.64+4.84+0.04+3.24+0.04)/5=1.76
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.mean(arr))
print(np.var(arr)) 
print(np.std(arr)) #sqrt(var)
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.mode(arr)) 
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.prod(arr)) 
```

```python
print(arr)
print(np.dot(arr)) 
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.sum(arr,axis=0))
```

```python
arr = np.random.randint(1,10,5)
print(arr)
print(np.sum(arr,axis=1))
```

```python
##### Note Axis=0 Row wise operation but column wise result
arr1 = np.random.randint(1,10,5)
arr2 = np.random.randint(1,10,5)
print(arr1)
print(arr2)
print(np.sum((arr1,arr2),axis=0))
```

```python
##### Note Axis=1 Column wise operation but row wise result
print(arr1)
print(arr2)
print(np.sum((arr1,arr2),axis=1))
```

```python
arr1 = np.random.randint(1,10,5)
arr2 = np.random.randint(1,10,5)
print(arr1)
print(arr2)
print(np.average((arr1,arr2),axis=0))
```

```python
print(arr1)
print(arr2)
print(np.average((arr1,arr2),axis=1))
```

```python
arr1 = np.random.randint(1,10,5)
arr2 = np.random.randint(1,10,5)
print(arr1)
print(arr2)
print(np.prod((arr1,arr2),axis=0))
```

```python
arr1 = np.random.randint(1,10,5)
arr2 = np.random.randint(1,10,5)
print(arr1)
print(arr2)
print(np.prod((arr1,arr2),axis=1))
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([1,2,3])
print(arr1)
print(arr2)
print(np.dot(arr1,arr2))
print(np.prod((arr1,arr2)))
```
