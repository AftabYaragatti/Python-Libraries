##### **2.2 var_name.describe()**
##### It returns the statistical information includes count, mean ,standarad deviation ,min ,max, 25%,50%,75%.
##### **Syntax : var_name.describe(percentiles=None, include=None, exclude=None)**
##### percentiles ,include and exclude is optional

```python
import pandas as pd
import numpy as np
s = pd.Series(data=np.random.randint(30,70,10),index=np.arange(10),name='Marks')
s.to_frame()
```

##### Before Sorting the Values are : [50,48,44,47,49,56,64,62,34,40]
##### After sorting values are : [34,40,44,47,48,49,50,56,62,64] => (Index: 0 → 9)
##### Calculate 25% (Q1)
##### **Formula = (total_elements - 1) * 25%**
##### = (10 - 1) * 0.25 = 9 * 0.25 = 2.25
##### Position = 2.25
##### Index 2 → 44
##### Index 3 → 47
##### Interpolation:
##### value = 44 + (47 - 44) * 0.25
##### = 44 + 3 * 0.25
##### = 44 + 0.75
##### = 40.75
##### **25% => 44.75000**
##### **(Median) Calculate 50% (Q2)**
##### **Formula = pos = (total_elements - 1) * 50%**
##### = (10 - 1) * 0.5 = 9 * 0.5 = 4.5
##### Position = 4.5
##### Index 4 → 48
##### Index 5 → 49
##### Interpolation:
##### value = 48 + (49 - 48) * 0.5
##### = 48 + 1 * 0.5
##### = 48 + 0.5
##### = 48.50
##### **50% => 48.50000**
##### **Calculate 75% (Q3)**
##### **Formula = pos = (total_elements - 1) * 75%**
##### = (10 - 1) * 0.75 = 9 * 0.75 = 6.75
##### Position = 4.5
##### Index 6 → 50
##### Index 7 → 56
##### Interpolation:
##### value = 50 + (56 - 50) * 0.75
##### = 50 + 6 * 0.75
##### = 50 + 4.5
##### = 54.50
##### **75% => 54.50000**

```python
s.describe()
```

##### We are telling Pandas to show only the 25% percentile (Q1) instead of default [0.25, 0.5, 0.75]. But pandas always shown 50% default

```python
s.describe(percentiles=[0.25])
```

```python
s.describe(percentiles=[0.25,0.75])
```

```python
s.describe(percentiles=[0.25,0.5,0.75])
```

```python
s.describe(percentiles=[0.1,0.9])
```

##### 3. values = Similar to index, but uses actual numeric values
##### What method='values' means : Uses the actual index VALUES (numeric) to calculate interpolation
##### It treats your Series like (x, y) points
##### x → index
##### y → values
##### For a missing value between two points: y = y1 + ( (x - x1) / (x2 - x1) ) * (y2 - y1)
##### Example : We want value at index 3
##### Known points:
##### x1 = 1 → y1 = 10
##### x2 = 5 → y2 = 30
##### x  = 3 → ?
##### Apply formula:
##### y = 10 + ((3 - 1) / (5 - 1)) * (30 - 10)
##### y = 10 + (2/4) * 20
##### y = 10 + 10
##### y = 20
##### ✅ Output
##### 1    10.0
##### 3    20.0
##### 5    30.0

```python
import pandas as pd
import numpy as np
s=pd.Series(data=[14, np.nan,28,np.nan,34], index=[7,2,1,9,3],name='Sample')
s.to_frame()
```

```python
s.interpolate(method='values')
```
