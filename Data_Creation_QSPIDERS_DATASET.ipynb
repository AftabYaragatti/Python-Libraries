# **CREATED PYSPIDERS DATA SET**

```python
import pandas as pd
import numpy as np
```

```python
n=30000
```

```python
Student_Name = ['Stud_'+str(i).zfill(5) for i in range(1,n+1)]
```

```python
pd.DataFrame(Student_Name,columns=['Student Name'])
```

```python
Phone_Num = np.random.randint(6000000000,9999999999,n,dtype=np.int64)
```

```python
pd.DataFrame(Phone_Num,columns=['Phone Number'])
```

```python
Email_ID = [i+'@gmail.com' for i in Student_Name]
```

```python
pd.DataFrame(Email_ID,columns=['Email ID'])
```

```python
Admission_Type = np.random.choice(['INC','PAID'],n)
```

```python
pd.DataFrame(Admission_Type,columns=['Admission Type'])
```

```python
Tenth_Percentage = np.round(np.random.randint(60,99,n) + np.random.rand(n),2)
```

```python
pd.DataFrame(Tenth_Percentage,columns=['Tenth Percentage'])
```

```python
Tenth_YOP = np.random.choice([2017,2018,2019,2020],n)
```

```python
pd.DataFrame(Tenth_YOP,columns=['Tenth YOP'])
```

```python
Twelth_Percentage = np.round(np.random.randint(60,99,n) + np.random.rand(n),2)
```

```python
pd.DataFrame(Twelth_Percentage,columns=['Twelth Percentage'])
```

```python
Twelth_YOP = Tenth_YOP+2
```

```python
pd.DataFrame(Twelth_YOP,columns=['Twelth YOP'])
```

```python
Degree_Percentage = np.round(np.random.randint(60,99,n) + np.random.rand(n),2)
```

```python
pd.DataFrame(Degree_Percentage,columns=['Degree Percentage'])
```

```python
Branches = np.random.choice(['Bsc', 'Bcom', 'BE', 'BTech', 'BBA', 'BCA'],n)
```

```python
pd.DataFrame(Branches,columns=['Branches'])
```

```python
Degree_YOP = []
for branch , yop in zip(Branches,Twelth_YOP):
    if branch in ['Bsc','Bcom','BBA','BCA']:
        Degree_YOP.append(yop+3)
    elif branch in ['BE','BTech']:
        Degree_YOP.append(yop+4)
Degree_YOP = np.array(Degree_YOP)
```

```python
pd.DataFrame(Degree_YOP,columns=['Degree YOP'])
```

```python
Master_Stream = np.random.choice(['ME','MTech','MCA',None],n)
```

```python
pd.DataFrame(Master_Stream,columns=['Master Stream'])
```

```python
Master_Degree_Percentage = []
for i in Master_Stream:
    if i!=None:
        Master_Degree_Percentage.append(np.random.randint(35,99))
    else:
        Master_Degree_Percentage.append(np.nan)
Master_Degree_Percentage = np.array(Master_Degree_Percentage)
```

```python
pd.DataFrame(Master_Degree_Percentage,columns=['Master Degree Percentage'])
```

```python
out = []
for i,j in zip(Master_Stream,Degree_YOP):
    if i in ['ME','MTech','MCA']:
        out.append(j+2)
    else:
        out.append(np.nan)
Master_YOP = np.array(out)
```

```python
pd.DataFrame(Master_YOP,columns=['Master YOP'])
```

```python
Courses_Taken = np.random.choice(['Python FullStack Development','Python Data Analytics','Python Data Science','Python Selenium'],n)
```

```python
pd.DataFrame(Courses_Taken,columns=['Courses Taken'])
```

```python
Status_Of_Placement = np.random.choice(['Placed','Not Placed'],n)
```

```python
pd.DataFrame(Status_Of_Placement,columns=['Status Of Placement'])
```

```python
Pyspider_Dataset = pd.DataFrame({
    'Student Name': Student_Name,
    'Phone Number' : Phone_Num,
    'Email ID': Email_ID,
    'Admission Type' : Admission_Type,
    'Courses Taken':Courses_Taken,
    'Tenth YOP': Tenth_YOP,
    'Tenth Percentage':Tenth_Percentage,
    'Twelth YOP':Twelth_YOP,
    'Twelth Percentage':Twelth_Percentage,
    'Degree Yop':Degree_YOP,
    'Degree Stream':Branches,
    'Degree Percentage':Degree_Percentage,
    'Masters Degree YOP':Master_YOP,
    'Masters Degree Stream':Master_Stream,
    'Masters Degree Percentage':Master_Degree_Percentage,
})
```

```python
Pyspider_Dataset
```

```python
Pyspider_Dataset.to_csv("Pyspider_Dataset.csv", index=False)
```

```python
import pandas as pd
import random
from datetime import datetime

CURRENT_YEAR = 2026

def generate_joining_date(row):
    # Step 1: Get latest education year
    if pd.notna(row['Masters Degree YOP']):
        yop = int(row['Masters Degree YOP'])
    else:
        yop = int(row['Degree Yop'])
    
    # Step 2: Fix future year issue
    if yop > CURRENT_YEAR:
        yop = CURRENT_YEAR
    
    # Step 3: Generate valid joining date
    month = random.randint(1, 12)
    day = random.randint(1, 28)
    return datetime(yop, month, day).date()

# Apply to dataset
Pyspider_Dataset['Course_Enrollment_Date'] = Pyspider_Dataset.apply(generate_joining_date, axis=1)
```

```python
Pyspider_Dataset.to_csv("Pyspider_Dataset.csv", index=False)
```
