##### **1.1 Column Access :**
##### It is used to access one or more columns.
##### Syntax For single column (Series) - **df['col']**
##### Syntax For multiple columns (DataFrame) - **df[['col1','col2']]**

```python
import pandas as pd
import numpy as np
df=pd.DataFrame(data={'Student_ID':[101,102,103,104,105],'Names':['Aftab','Smith','Scott','Bob','Jhon'],'Marks1':np.arange(70,95,5),
                      'Marks2':np.arange(50,75,5),'Marks3':np.arange(40,65,5),'Marks4':np.arange(75,100,5),'Marks5':np.arange(65,90,5),
                      'Marks6':np.arange(70,95,5)},index=np.arange(1,6))
df
```

##### I need to access the Names column only (like series)
##### **To get single Column we uses df[column_name]**

```python
df['Names']
```

##### I need to access the Multiple columns such as Names and Student_ID column only (like Pandas)
##### **To get Multiple Columns we uses df[['column_name1','column_name2']]**

```python
df[['Names','Student_ID']]
```

##### **To Change the order of columns we uses df[['column_name3','column_name1','column_name2']]**

```python
df[['Names','Marks2','Student_ID']]
```

#### **1.2 Row Access :**

##### **1.2.1 Using loc (Label-based) :**
##### Uses index labels to get the rows
##### **Genral syntax :** df.loc[row_label]
##### **Slicing Syntax :** df.loc[start_row:end_row]                      
##### **Special Value Syntax :** df.loc[Row_index, 'col_name']           
##### **Accessing multiple Rows :** df.loc[row_index_start:row_index_end, ['col_name1','col_name2']]

```python
import pandas as pd
import numpy as np
df=pd.DataFrame(data={'Student_ID':[101,102,103,104,105],'Names':['Aftab','Smith','Scott','Bob','Jhon'],'Marks1':np.arange(70,95,5),
                      'Marks2':np.arange(50,75,5),'Marks3':np.arange(40,65,5),'Marks4':np.arange(75,100,5),'Marks5':np.arange(65,90,5),
                      'Marks6':np.arange(70,95,5)},index=np.arange(1,6))
df
```

##### If you want to get the particular row we need to use loc[] Where it works based on index label
##### **To get single row we uses df.loc[row_label]**
##### **Genral syntax :** df.loc[row_label]

```python
df.loc[1]
```

##### If you want multiple rows we need to use nested list [[]] and separated each row label by comma.

##### **To get Multiple row we uses df.loc[['row_label1','row_label2','row_label3']]**

```python
df.loc[[1,3,5]]
```

##### **1.4 Conditional Filtering :**
##### It is used to acess the data by condition.
##### **General Syntax :** df[condition]
##### **Example 1 :** df[(df['Age'] > 20) & (df['Salary'] > 3000)]
##### **Example 2 :** df[df['Name'] == 'A']

```python
import pandas as pd
import numpy as np
df=pd.DataFrame(data={'Student_ID':[101,102,103,104,105],'Names':['Aftab','Smith','Scott','Bob','Jhon'],'Marks1':np.arange(70,95,5),
                      'Marks2':np.arange(50,75,5),'Marks3':np.arange(40,65,5),'Marks4':np.arange(75,100,5),'Marks5':np.arange(65,90,5),
                      'Marks6':np.arange(70,95,5)},index=np.arange(1,6))
df
```

```python
df[df['Marks1']>80]
```

```python
df[(df['Marks1']>70) & (df['Marks1']<90)]
```

```python
df[(df['Marks1']>70) | (df['Marks1']<90)]
```

```python
df[~(df['Marks1']<90)]
```

```python
df[(df['Names']=='Scott')]
```

```python
df[~(df['Names']=='Scott')]
```

```python
df[(df['Names']=='Aftab') | (df['Marks1']<90)]
```

```python
df[(df['Names']=='Aftab') | (df['Marks1']>80)]
```
