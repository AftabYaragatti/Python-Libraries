##### **Note :** If we want to work with Multiple Series then we must concat() them to make it Multiple Sereis().

#### **Attributes for Multiple Pandas Series()**
##### 1.1 var_name.index` → labels/indexes - It returns indexes of Multiple Series(DataFrame)
##### 1.2 var_name.values` → data (NumPy array) - It returns all the values in row and column wise (2D array form)
##### 1.3 var_name.dtypes` → data type - It returns the data types with respect to column name
##### 1.4 var_name.shape` → size (tuple) - It returns the shape of the array (no.of rows,no.of columns)
##### 1.5 var_name.size` → number of elements - It returns total number of elements
##### 1.6 var_name.ndim` → dimension (always 1) - It returns dimension of the array (2D)
##### 1.7 var_name.columns` → name of Series - It returns all the column names
##### 1.8 var_name.axes` → list of index - It returns the indexs of columns and column names.

```python
import pandas as pd
import numpy as np
df1 = pd.Series(data=np.random.randint(60,90,5),index=[1,4,5,2,6],name='Sales1')
df2 = pd.Series(data=np.random.randint(60,90,5),index=[1,4,5,2,6],name='Sales2')
df3 = pd.Series(data=np.random.randint(60,90,5),index=[1,4,5,2,6],name='Sales3')
df4 = pd.Series(data=np.random.randint(60,90,5),index=[1,4,5,2,6],name='Sales4')
df5 = pd.Series(data=np.random.randint(60,90,5),index=[1,4,5,2,6],name='Sales5')
df = pd.concat([df1,df2,df3,df4,df5],axis=1)
df
```

```python
df.index
```

```python
df.values
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.size
```

```python
df.ndim
```

```python
df.columns
```

```python
df.axes
```
