#### **1.1 Indexing :**

##### It is the process of giving an subaddress for all elements present in a collection.Indexing is used to fetch single element or update a single element in array

##### **1.1.1(a) : 1-Dimensional Indexing :**
##### **Syntax for accessing : arr_var[index]**
##### **Syntax for Update : arra_var[index]=new_value**

##### **Example 1:**

```python
import numpy as np
arr = np.arange(11,21)
arr
```

##### **Positive Indexing :** starts from 0 to lenght(array)-1

```python
# I want to access element 11
print(arr[0])
print(arr[0].item())
print(int(arr[0]))
```

```python
# I want to access element 15 & 19
print(arr[4])
print(arr[8])
```

```python
# I want to update element 11 to 100 and 16 to 150
arr[0]=100
arr[5]=150
print(arr)
```

##### **Negative Indexing :** starts from -1 to 1-lenght(array)

##### Accessing Array Elements based on Index

#### **1.1.2 Boolean Indexing :**

##### It is the process of fetching the elements based on given conditions
##### **Syntax : array_variable[condition]**
##### If condition True then only it **returns the array elements** 
##### It always returns an single 1D array elements or flatten array elements

##### **1.1.2(a) : 1-Dimensional Boolean Indexing**

```python
# Single Conditions
import numpy as np
arr = np.arange(10,20,2)
print(f'Orignal Array : {arr}')
print(f'After Condition Array : {arr[arr>15]}')
```

```python
# Multiple Conditions
import numpy as np
arr = np.arange(10,50,2)
print(f'Orignal Array : \n{arr}')
print(f'After Condition Array : {arr[(arr > 20) & (arr < 30)]}')
print(f'After Condition Array : {arr[(arr > 20) | (arr < 30)]}')
print(f'After Condition Array : {arr[~(arr > 20)]}')
```

##### **1.1.2(b) : 2-Dimensional Boolean Indexing**

```python
# Single Conditions
import numpy as np
arr = np.arange(10,20).reshape(2,5)
print(f'Orignal Array : {arr}')
print(f'After Condition Array : {arr[arr>15]}')
```

```python
# Multiple Conditions
import numpy as np
arr = np.arange(10,20).reshape(2,5)
print(f'Orignal Array : {arr}')
print(f'After Condition Array : {arr[(arr > 15) & (arr < 18)]}')
print(f'After Condition Array : {arr[(arr > 15) | (arr < 18)]}')
print(f'After Condition Array : {arr[~(arr > 15)]}')
```

##### **1.1.2(c) : 3-Dimensional Boolean Indexing**

```python
# Single Conditions
import numpy as np
arr = np.arange(10,30).reshape(5,2,2)
print(f'Orignal Array : {arr}')
print(f'After Condition Array : {arr[arr>15]}')
```

```python
# Multiple Conditions
import numpy as np
arr = np.arange(10,30).reshape(5,2,2)
print(f'Orignal Array : {arr}')
print(f'After Condition Array : {arr[(arr > 15) & (arr < 18)]}')
print(f'After Condition Array : {arr[(arr > 15) | (arr < 18)]}')
print(f'After Condition Array : {arr[~(arr > 15)]}')
```

#### **1.1.3 Fancy Indexing :**

##### It is the process of fetching the elements based on group of indexs
##### **Syntax : array_variable[[index1],[index2],[index3]....]**
##### It always returns an single 1D array elements or flatten array elements
##### We need to create two array 1 is for creating entire array and another one is for giving indexes to the created entire array (or we can directly give the indexes)
##### It depends how we are passing array indexs whether we are passing in 1D [] or 2D [[]] or 3D [[[]]]

#### **1.1.3(a) 1-Dimensional Fancy Indexing :**
##### If we are returning an array in 1D means we should use just [] square brackets
##### If we are retruning 2D means [[],[]] or [[index1,index2]] 
##### If we are returning 3D means [[[]]] or [[[],[]],[[],[]]]
##### If we are passing multiple arr_indexes Then it must be [[index1,index2]] form or [[[index1,index2]]]

```python
# 1D array with 1D array index , that returns 1D Array
arr1 = np.arange(10,20)
print(f'Original Array : {arr1}')
arr2 = [-1,-2,-3,-4,-5] #indexes of arr1
arr = arr1[arr2]
print(f'1-D Array : {arr}')
print(arr.shape)
```

```python
# 1D array with 1D array index , that returns 2D Array
arr1 = np.arange(10,20)
print(f'Original Array : {arr1}')
arr2 = [-1,-2,-3,-4,-5] #indexes of arr1
arr = arr1[[arr2]]
print(f'2-D Array : \n{arr}')
print(arr.shape)
arr = arr1[[arr2,arr2]]
print(f'2-D Array : \n{arr}')
print(arr.shape)
```

```python
# 1D array with 1D array index , that returns 2D Array
arr1 = np.arange(10,20)
print(f'Original Array : {arr1}')
arr2 = [-1,-2,-3,-4,-5] #indexes of arr1
arr = arr1[[arr2],[arr2]] #Error not possible
print(f'2-D Array : \n{arr}')
print(arr.shape)
arr = arr1[arr2,arr2] # Error not possible
print(f'2-D Array : \n{arr}')
print(arr.shape)
```

```python
# 1D array with 1D array index , that returns 3D Array
arr1 = np.arange(10,20)
print(arr1)
arr2 = [-1,-2,-3,-4,-5] #indexes of arr1
arr = arr1[[[arr2]]]
print(f'3-D Array : \n{arr}')
print(arr.shape)
arr = arr1[[[arr2],[arr2]]]
print(f'3-D Array : \n{arr}')
print(arr.shape)
arr = arr1[[[arr2,arr2],[arr2,arr2]]]
print(f'3-D Array : \n{arr}')
print(arr.shape)
```

```python
# 1D array with 1D array index , that returns 1D,2D & 3D Array
arr1 = np.arange(10,20)
print(arr1)
arr2 = [0,-1] #indexes of arr1
arr3 = [1,-2] #indexes of arr1
arr_11 = arr1[arr2] #returns 1D array
print(f'1D Array : \n{arr_11}')
print(arr_11.shape)
arr_12 = arr1[arr3] #returns 1D array
print(f'1D Array : \n{arr_12}\n')
print(arr_12.shape)
arr_2 = arr1[[arr2,arr3]] #returns 2D array
print(f'2D Array : \n{arr_2}\n')
print(arr_2.shape)
arr_3 = arr1[[[arr2,arr3]]] #returns 3D array
print(f'3D Array : \n{arr_3}\n')
print(arr_3.shape)
```

```python
arr1 = np.arange(10,20)
print(arr1)
arr2 = [-1,-2,-3,-4,-5] #indexes of arr1
arr3 = [-6,-7,-8,-9,-10] #indexes of arr1
arr_1 = arr1[arr2] #returns 1D array
arr_2 = arr1[[arr2,arr3]] #returns 2D array
arr_3 = arr1[[[arr2,arr3]]] #returns 3D array
print(f'1D Array : \n{arr_1}\n2D Array : \n{arr_2}\n3D Array : \n{arr_3}')
print(arr_1.shape)
print(arr_2.shape)
print(arr_3.shape)
```

```python
arr1 = np.arange(10,20)
print(f'Original : {arr1}')
arr2 = [-1,-2,-3] #indexes of arr1
arr3 = [-4,-5,-6] #indexes of arr1
arr4 = [-7,-8,-9] #indexes of arr1
arr = arr1[[arr2,arr3,arr4]] #returns 3D array
print(arr)
```

#### **1.2 Slicing :**

##### It is the process of fetching elements in sequencial order

##### **Syntax for 1D Slicing : arr_var[start:stop:step]**
##### **Syntax for 2D Slicing : arr_var[start_row:stop_row:step_row , start_col:stop_col:step_col]**
##### **Syntax for 3D Slicing : arr_var[start_row:stop_row:step_row , start_col:stop_col:step_col , start_ele:stop_ele:step_ele,]**

#### **1.2.1 : 1-Dimensional Slicing**

```python
import numpy as np
arr = np.random.randint(1,100,20)
print(arr)
```

```python
# I need to access elements 20 to 77
print(arr[1:16])
```

```python
# I need to access elements 20,17,40,28,86,68,36 & 77
print(arr[1:16:2])
```

```python
# I need to access elements 77,36,68,86,28,40,17,20
print(arr[15::-2])
```

```python
# I need to access alternative elements 
a=arr[::2]
print(arr[::2])
print(a.size)
```

##### To Understand it we need to trace 

```python
print(f'Original Array : {arr}')
# I need to acces 8 ,7 ,20
print(f'Access Start Index : {arr[6]}')
print(f'Access Stop Index : {arr[6:10+1]}')
print(f'Access Step Index : {arr[6:11:2]}')
```

#### **1.2.2 : 2-Dimensional Slicing**

```python
import numpy as np
arr = np.random.randint(1,100,20).reshape(4,5)
print(arr)
```

```python
# I need to access the element 
# [41,75,41]
# [47,9,65]
print(arr[0:3:2,0::2])
```

```python
# I need to access the element 
# [11,79]
# [36,72]
print(arr[1::2,1::2])
```

```python
# I need to access the element 
# [41,2]
# [65,88]
print(arr[0::2,-1:-5:-3])
```

```python
# I need to access the element 
# [78,73]
# [6,79]
# [73,89]
print(arr[-1::-2,-1::-2].T)
```

```python
# I need to access the element 
# [41,65]
# [75,9]
# [41,47]
print(arr[0::2,-1::-2].T)
```

#### **1.2.3 : 3-Dimensional Slicing**

```python
import numpy as np
arr = np.random.randint(1,100,24).reshape(2,3,4)
print(arr)
```

```python
# Access the following elements from the array
# [[45,72]
# [37,79]]
# [[21,73]
# [46,16]]
print(arr[0:,0::2,0::2])
```

#### **1.3 : All()**
##### It is a numpy method which is used to check the condition and values.
##### If all the conditions or values are True Then only it returns True.
##### If any one condition or value are False Then it returns False.
##### **Syntax : np.all(condition)**

```python
import numpy as np
arr = np.random.randint(1,100,10)
print(arr)
```

```python
print(np.all(arr>1))
```

```python
print(np.all(arr>0))
```

```python
print(np.all(arr%2==0))
```

```python
arr = np.arange(1,20,2)
print(np.all(arr%2!=0))
```

```python
arr = np.arange(0,20,2)
print(np.all(arr%2==0))
```

```python
arr = np.arange(0,20,2)
print(np.all(arr%2!=0))
```

#### **1.4 : Any()**
##### It is a numpy method which is used to check the condition or values is True or Not.
##### If any one condition or value are True Then it returns True.
##### If all the conditions or values are False Then it returns False.
##### **Syntax : np.any(condition)**

```python
import numpy as np
arr = np.random.randint(1,100,10)
print(arr)
```

```python
print(np.any(arr>1))
```

```python
print(np.any(arr<1))
```

```python
print(np.any(arr%2==0))
```

```python
print(np.any(arr%2!=0))
```

#### **1.5 : Where()**
##### It is used to fetch the values based on given conditions
##### By using where we fill the values according to the condition and those values must be scalar
##### **Syntax : np.where(condition)** - It return Indexs of that elements where condtion is satisfies
##### **Syntax : np.where(condition,TSB,FSB)** - It returns Values or Fills Values Based on Conditions
##### Where TSB-True Statement Block & FSB-False Statement Block is Optional

```python
import numpy as np
arr = np.random.randint(1,100,10)
print(arr)
```

```python
print(np.where(arr%2==0))
```

```python
print(np.where(arr%2==0,1,0))
```

```python
print(np.where(arr%2==0,arr,0))
```
