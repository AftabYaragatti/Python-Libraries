##### **1.1 astype()**
##### **Syntax : var_name.astype(dtype, copy=True, errors='raise')**
##### **Parameters:**
##### dtype → target datatype (int, float, str, etc.)
##### copy → return new object (default True)
##### errors → 'raise' or 'ignore'

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name='Marks')
df.to_frame()
```

```python
df.astype(dtype=float)
```

##### **1.2 tolist()**
##### Convert Series → Python list
##### **Syntax : var_name.tolist()** or
##### **Syntax : var_name.to_list()**

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name='Marks')
df.to_frame()
```

```python
df.to_list()
```

##### **1.3 to_numpy()**
##### Convert to numpy array
##### **Syntax : var_name.to_numpy(dtype=None, copy=False, na_value=None)**
##### dtype,copy and na_value are optional

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name='Marks')
df.to_frame()
```

##### Data is converted into numpy array

```python
df.to_numpy()
```

```python
import pandas as pd
import numpy as np
df = pd.Series(data=[12,90,None,98,None],index=np.arange(5),name='Marks')
df.to_frame()
```

##### NaN values converted to 0 if na_value is False

```python
df.to_numpy(na_value=False)
```

##### NaN values converted to 0 if na_value is True

```python
df.to_numpy(na_value=True)
```

##### **1.5 copy()**
##### Creates copy of object
##### **Syntax : var_name.copy()**

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name='Marks')
df.to_frame()
```

```python
d=df.copy()
```

```python
d.to_frame()
```

```python
df.to_frame()
```

```python
d[0]=100
d.to_frame()
```

```python
df.to_frame()
```

##### **1.6 to_frame(name=None)()**
##### CConvert Series → DataFrame
##### **Syntax : var_name.to_frame(name=None)**

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name='Marks')
df.to_frame()
```

##### **1.8 to_json()**
##### Convert to JSON format
##### **Syntax : var_name.to_json(path_or_buf=None)**

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name='Marks')
df.to_frame()
```

##### If we specified the path then, We will get json form data to that file Hello_json.json file

```python
df.to_json('Hello_Json.json')
```

##### If json file is not present , it will created a json file and stores the json data.

```python
df.to_json('JSON.json')
```

##### **1.9 to_csv()**
##### Convert to CSV format
##### **Syntax : var_name.to_csv("data.csv", index=False)**

```python
import pandas as pd
import numpy as np
df = pd.Series(data=np.random.randint(50,80,5),index=np.arange(5),name='Marks')
df.to_frame()
```

```python
df.to_csv()
```

##### If csv file is not present , it will created a csv file and stores the csv data. By default index value is True

```python
df.to_csv('data.csv')
```

##### If index=False means it will not store the index values in csv file

```python
df.to_csv('data1.csv',index=False)
```
