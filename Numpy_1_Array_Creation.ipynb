##### **============================================================** 
##### **1.1.4.Priority in Numpy :**
##### While creating any type of array , If we pass entire array elements as as int then it will display array elements as it is....
##### If we pass any one array element as as float or complex then it will display array elements based on the **Order of Priority**....

##### **Order of Priority from Lowest Priority to Highest Priority :**
##### 1.bool
##### 2.int
##### 3.float
##### 4.Complex
##### 5.string
##### 6.object

##### **1.1.5.Normal Array :** Entire array elements are integer type only

```python
import numpy as np
arr = np.array([1,2,3,4,5])
print(arr)
```

##### **1.1.6.Mixed Data type Array :** (int + float)
##### Suppose if I am passing integer elements and float elements then, 
##### It will returns the array elements based on the Order of Priority :::::::

```python
import numpy as np
arr = np.array([1 , 2 , 3.6 , 4 , 5.8])
print(arr) 
```

##### **Note :** We will get float elements in array, BECAUSE FLOAT HAVING HIGHEST PRIORITY THAN INT

##### **1.1.7.Mixed Data type Array :** (int + float + complex)
##### Suppose if I am passing integer,float and complex elements then, 
##### It will returns the array elements based on the Order of Priority :::::::

```python
import numpy as np
arr = np.array([1+9j , 2 , 3 , 4 , 5])
print(arr) 
```

##### **Note :** We will get float elements in array with complex elements,BECAUSE FLOAT HAVING HIGHEST PRIORITY THAN INT AND COMPLEX HAVING HIGHER PRIORITY THAN FLOAT

##### **1.1.8.Mixed Data type Array :** (int + float + complex + string)

```python
import numpy as np
arr = np.array([4+7j , 2 , 3.8 , 4 , "5"])
print(arr)
```

##### **NOTE :** Here In example 4 we passed integer elements as well as float elements and string elements and String Has Highest priority than int and float, But here it will conver the entire array list elements to the string only with considering int and float and complex also directly converted to the string as it is

##### **1.1.9.Mixed Data type Array :** (bool + int)

#### **1.2 np.arange():**
##### **============================================================** 
##### This is an inbuilt NumPy function that returns evenly spaced values within a given interval.
##### **Syntax :** **np.arange( start , stop, step , dtype=None )**
##### It returns array elements interms of Float as a Default array data type
##### dtype is optional here
##### start value is also optional if you dont want to specify, but if you want specific start value we can use start value. Similar step value is optional, If we want specific step we can use it
##### **Parameters:**
##### **start** - starting value of array(including starting value) 
##### **stop** - ending value of array(excluding ending value)
##### **step** - step is used to increment or decrement step value
##### **dtype** - used to change the type of data and convert the data from one data type of array to other type of array
##### default Data type (dtype) is **Float**

##### **start & stop** : We need to speicify the start and stop value while creating an array
##### start value is not manadatory, if we doesn't specify it will consider default value as 0

```python
# 1D Array :
import numpy as np
arr=np.arange(10)
print(arr)
print(type(arr))
print(arr.dtype)
```

```python
# 1D Array :
import numpy as np
arr=np.arange(1,10)
print(arr)
print(type(arr))
print(arr.dtype)
```

##### **start,stop & step** : We need to speicify the start and stop value while creating an array
##### start value is not manadatory, if we doesn't specify it will consider default value as 0
##### But if we doesn't specify start value directly we specify step value means the interpreter will treat as start and stope value and returns empty array
##### By Default step value is 1 always

```python
# 1D Array :
import numpy as np
arr=np.arange(1,10,2)
print(arr)
print(type(arr))
print(arr.dtype)
```

#### **1.3. np.ones():**
##### **============================================================** 
##### This function is used to get a new array of given shape and type filled with ones (1).
##### **Syntax:** **np.ones(shape, dtype=None, order='C')**
##### Default data type is float
##### dtype and order is optional
##### **Parameters:**
##### **shape** - Used for dimension,It gives shape for array to create any(1D,2D,3D) dimension of array.
##### shape(for 1D) - [rows]
##### shape(for 2D) - [rows,cols]
##### shape(for 3D) - [rows,cols,n(elements)]
##### **dtype** - Used to convert the type of data from one data type to another data type
##### **order** - Used to store the array elements in a memory
##### The order="C" parameter defines how the array is stored in memory, specifically using Row-Major Order(Row by Row). 
##### The order="F" parameter defines how the array is stored in memory, specifically using Column-Major Order(Column by Column). 

##### **1D Array :** If we want to create 1 Dimension array just we need to specify the single element. 
##### we can declare array elements size inside the square brackets[] or paranthesis () or keep as it is

##### By default it returns array elements as floating data type
##### For 1D Array we can pass directly size of array (n) , and you can pass n value inside [] or ()   

```python
# 1D Array :
import numpy as np
arr=np.ones(4)
print(arr)
print(arr.dtype)
```

```python
# 1D Array :
import numpy as np
arr=np.ones((4))
print(arr)
print(arr.dtype)
```

```python
# 1D Array :
import numpy as np
arr=np.ones([4])
print(arr)
print(arr.dtype)
```

##### **axis** : When You are working with multi-dimensional data([1,2,3],[5,4,3]) and we want control over how interpolation is arranged
##### If start and stop values are scalars (numbers), Then axis does NOT matter,It Always gives a 1D array
##### If start and stop values are arrays (vectors), Then axis controls how output is structured
##### axis = 0 (default) → values grow row-wise (downwards) 
##### axis = 1 → values grow column-wise (sideways)
##### For axis = 0 
##### => [ [start]
#####    [middle]
#####     [stop] ]
##### For axis = 1
##### => [ start → middle → stop ]
##### we can't pass more than 2 arrays...... because 1st array takes as start value and 2nd array takes as stop value as an arguement
##### If we pass more than 2 the third array takes num value as an 3rd arguemnet and throws error

##### **For axis=0 :**

##### Second row values starts from row2 and stop value col2(start=row2,stop=col2)

##### **============================================================** 
#### **1.9 np.empty() :** 
##### **============================================================** 
##### It creates a new array with old existing values from our main memory(Garbage Memory).
##### **Syntax : np.empty(shape, dtype=None, order='C')**
##### It generating the garbage values from main memory.
##### datatype(returns float data type) and order is optional here 
##### Its manadatory to pass shape of array and specific shape value to create the entire array
##### **Parameters:**
##### shape (tuple of int): Shape of the new array (1D/2D/3D).
##### For 1D - np.empty(n)
##### For 2D - np.empty(rows,cols)
##### FOr 3D - np.empty(rows,cols,n)
##### dtype : data type of the array.
##### order ({'C', 'F'} : Stroing array elements in an memory

##### **1-Dimensional Array :** We Need to pass number of elements

```python
import numpy as np
arr = np.empty(4)
arr
```

##### **2-Dimensional Array :** We Need to pass number of rows and columns

```python
import numpy as np
arr = np.empty((3,4))
arr
```

##### **3-Dimensional Array :** We Need to pass number of rows and columns with elements

```python
import numpy as np
arr = np.empty((2,2,3))
arr
```

##### **1-Dimensional with dtype :**

```python
import numpy as np
arr = np.empty(4,dtype=int)
print(arr,arr.dtype)
arr = np.empty(4,dtype=float)
print(arr,arr.dtype)
arr = np.empty(4,dtype=complex)
print(arr,arr.dtype)
arr = np.empty(4,dtype=bool)
print(arr,arr.dtype)
arr = np.empty(4,dtype=str)
print(arr,arr.dtype)
arr = np.empty(4,dtype=list)
print(arr,arr.dtype)
arr = np.empty(4,dtype=tuple)
print(arr,arr.dtype)
arr = np.empty(4,dtype=set)
print(arr,arr.dtype)
arr = np.empty(4,dtype=frozenset)
print(arr,arr.dtype)
arr = np.empty(4,dtype=dict)
print(arr,arr.dtype)
arr = np.empty(4,dtype=range)
print(arr,arr.dtype)
```

##### **order** : used to store array elements in a memory

```python
arr = np.empty((2,3),dtype=int,order="C")
print(arr,arr.dtype)
```

##### **============================================================** 
##### It generates identity matrix by default and It performs n*n matrix operation
##### **Syntax : np.identity(n , dtype=None)**

##### **Parameter:**

##### dtype : data type of the array.
