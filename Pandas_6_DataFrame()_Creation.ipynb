#### **1.DataFrame()**
##### Used to create a DataFrame from different data sources like list, dictionary, NumPy array, or Series.
##### **Syntax : pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)**
##### **Parameters :**
##### data → input data (list, dict, array, Series)
##### index → row labels
##### columns → column names
##### dtype → data type
##### copy → copy data or not

#### **OR**

##### **Syntax : pd.DataFrame(data, index, columns)** 
##### **Parameters:**
##### **data:** It is a dataset from which a DataFrame is to be created. It can be a list, dictionary, scalar value, series, and arrays, etc. 
##### **index:** It is optional, by default the index of the DataFrame starts from 0 and ends at the last data value(n-1). It defines the row label explicitly. 
##### **columns:** This parameter is used to provide column names in the DataFrame. If the column name is not defined by default, it will take a value from 0 to n-1. 
##### **Returns:** DataFrame object

##### **1.3 Create a DataFrame from Dictionary :**
##### Creating DataFrame from dict of ndarray/lists : We can create a DataFrame from a dictionary where the keys are column names and the values are lists or arrays.
##### -> All arrays/lists must have the same length.
##### -> If an index is provided, it must match the length of the arrays.
##### -> If no index is provided, Pandas will use a default range index (0, 1, 2, …).

```python
# -> All arrays/lists must have the same length.
import pandas as pd
data = {'Name':['Aftab','Smith','Goudra','Scott','Jhon'],'Marks':[94,89,78,91,94]}
pd.DataFrame(data)
```

```python
# -> If an index is provided, it must match the length of the arrays.
import pandas as pd
import numpy as np
data = {'Name':['Aftab','Smith','Goudra','Scott','Jhon'],'Marks':[94,89,78,91,94]}
pd.DataFrame(data,index=np.arange(10,60,10))
```

```python
# -> If no index is provided, Pandas will use a default range index (0, 1, 2, …).
import pandas as pd
import numpy as np
data = {'Name':['Aftab','Smith','Goudra','Scott','Jhon'],'Marks':[94,89,78,91,94]}
pd.DataFrame(data)
```

```python
import pandas as pd
import numpy as np
pd.DataFrame({'Name':['Aftab','Smith','Goudra','Scott','Jhon'],'Marks':[94,89,78,91,94]})
```

```python
import pandas as pd
import numpy as np
pd.DataFrame({'Name':['Aftab','Smith','Goudra','Scott','Jhon'],'Marks':[94,89,78,91,94]},index=[111,222,333,444,555])
```

##### **1.5 Create a DataFrame from Seires :**
##### We can create a DataFrame by using series also, where each series treated as rows and each series elements treated as Columns

```python
import pandas as pd
import numpy as np
s1=pd.Series(np.arange(10,60,10))
s2=pd.Series(np.arange(10,100+1,20))
s3=pd.Series(np.arange(100,600,100))
pd.DataFrame([s1,s2,s3])
```

##### **1.6 Create a DataFrame from a List of Dictionaries** 
##### We can create a DataFrame using a list of dictionaries, where each dictionary represents a row. This is useful for handling structured data from APIs or JSON, and is commonly used in web scraping and API processing.
##### We are storing the data like records if we use the list of dictionary.

```python
import pandas as pd
import numpy as np
data = [{'Name': 'Aftab', 'Degree': 'B.Tech', 'Score': 90},
        {'Name': 'Smith', 'Degree': 'BCA', 'Score': 84},
        {'Name': 'Jhon', 'Degree': 'B.Com', 'Score': 80} ]
pd.DataFrame(data)
```

```python
import pandas as pd
import numpy as np
data = [{'Name': 'Aftab', 'Degree': 'B.Tech', 'Score': 90},
        {'Name': 'Smith', 'Degree': 'BCA', 'Score': 84},
        {'Name': 'Jhon', 'Degree': 'B.Com', 'Score': 80} ]
pd.DataFrame(data,index=[101,102,103])
```

##### If we change the any column name or the column names are mismatched by ond dictionary to another dictionary it will take that column as a new column and rest of all the values will become NaN values.

```python
import pandas as pd
import numpy as np
data = [{'Name': 'Aftab', 'Course': 'B.Tech', 'Score': 90},
        {'Name': 'Smith', 'Degree': 'BCA', 'score': 84},
        {'name': 'Jhon', 'Degree': 'B.Com', 'Score': 80} ]
pd.DataFrame(data,index=[101,102,103])
```

##### **1.9 Create DataFrame using the zip() function**
##### get the list of tuples from two lists. and merge them by using zip().

```python
import pandas as pd
Name = ['tom', 'krish', 'nick', 'juli']
Age = [25, 30, 26, 22]
data = list(zip(Name, Age))
pd.DataFrame(data, columns=['Name', 'Age'])
```

##### **Index : It is used to change the default row index to specific row index**

```python
import pandas as pd
import numpy as np
pd.DataFrame({'Name':['Aftab','Smith','Goudra','Scott','Jhon'],'Marks':[94,89,78,91,94]},index=[111,222,333,444,555])
```

##### **Columns : It used to change the default column index to the specific column index**

##### **dtype : It is used to convert the one type of data into another type of data**

```python
import pandas as pd
import numpy as np
df=pd.DataFrame([['Aftab',98],['Smith',94],['Goudra',96],['Scott',89],['Jhon',89]],index=[111,222,333,444,555],columns=['Name','Marks'])
df
```

```python
pd.DataFrame(df['Marks'],dtype=float)
```

##### **copy : It is used to create a copy of the DataFrame, If True : It creates the copy and whatever changes we want to do we can, it will not affect the original DataFrame Series. If False : If we done any changes it directly affects to the original DataFrame**

```python
import pandas as pd
import numpy as np
df=pd.DataFrame([['Aftab',98],['Smith',94],['Goudra',96],['Scott',89],['Jhon',89]],index=[111,222,333,444,555],columns=['Name','Marks'])
df
```

```python
cp=pd.DataFrame(df['Marks'],dtype=float,copy=True)
cp
```

```python
cp['Marks']=cp['Marks']=100
```
