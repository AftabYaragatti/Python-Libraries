### **Copy() & View():**

#### **1.1 Copy():(Deep Copy)**
##### It is numpy method which is used to copy the complete array with different memory location. In copy the data will be independent (different address).
##### Creates a new independent array. Changes in copy DO NOT affect original
##### **Syntax : destination_variable = np.copy(array_variable)**
##### **NOTE : np. Means It will create a new array, .arr means it will modify original array itself**

```python
import numpy as np
original_arr = np.array([1,2,3,4,5,6,7])
copy_arr = np.copy(original_arr)
print(f'Original Array : {original_arr}')
print(f'Copied Array : {copy_arr}')
```

##### If I update with respect to copied array means it will not affect the original array

```python
import numpy as np
original_arr = np.array([1,2,3,4,5,6,7])
copy_arr = np.copy(original_arr)

print(f'Original Array : {original_arr}')
print(f'Copied Array : {copy_arr}')

copy_arr[-1]=500
copy_arr[0:2]=100,200

print()
print(f'Original Array : {original_arr}')
print(f'Copied Array : {copy_arr}')
```

```python
import numpy as np
arr1 = np.array([[1,2,3,4],[10,20,30,40]])
arr2 = np.copy(arr1)

print(f'Original Array : {arr1}')
print(f'Copied Array : {arr2}')

arr2[0][1]=120
arr2[1,2]=200

print()
print(f'Original Array : {arr1}')
print(f'Copied Array : {arr2}')

###### Array variable IDs address are different
print()
print(id(arr1))
print(id(arr2))

###### Array Elements address are same
print()
print(id(arr1[0,1]))
print(id(arr2[0,1]))

###### Array Varibale address are different
print()
print(arr1.ctypes.data)
print(arr2.ctypes.data)
```

#### **1.2 View():(Shallow Copy)**
##### It creates a replica of sarray with same address.
##### For objects address will be different but elements shares a same memory
##### Ot will not Creates a new independent array, It modifies same array with same memory address. Changes in copy it affect original
##### **Syntax : destination_variable = array_variable.view()**
##### **NOTE : np. Means It will create a new array, .arr means it will modify original array itself**

```python
import numpy as np
original_arr = np.array([1,2,3,4,5,6,7])
copy_arr = original_arr.view()
print(f'Original Array : {original_arr}')
print(f'Copied Array : {copy_arr}')
```

```python
import numpy as np
original_arr = np.array([1,2,3,4,5,6,7])
copy_arr = original_arr.view()
print(f'Original Array : {original_arr}')
print(f'Copied Array : {copy_arr}')
```

##### If we update with respect to original array. It will affect original memory too.

```python
import numpy as np
original_arr = np.array([1,2,3,4,5,6,7])
copy_arr = original_arr.view()

print(f'Original Array : {original_arr}')
print(f'Copied Array : {copy_arr}')

copy_arr[-1]=500
copy_arr[0:2]=100,200

print()
print(f'Original Array : {original_arr}')
print(f'Copied Array : {copy_arr}')
```

```python
import numpy as np
arr1 = np.array([[1,2,3,4],[10,20,30,40]])
arr2 = arr1.view()

print(f'Original Array : {arr1}')
print(f'Copied Array : {arr2}')

arr2[0][1]=120
arr2[1,2]=200

print()
print(f'Original Array : {arr1}')
print(f'Copied Array : {arr2}')
```

###### For Array Variable address are different because it is used shared memory but stored in different variable but for array elements address are same

```python
arr1=np.array([1,2,3,4])
arr2=np.array([1,2,3,4])
arr3=arr1.view()
print(id(arr1))
print(id(arr2))
print(id(arr3))
```

##### Even though ID's are giving differnt address but exact address is same by both address

```python
arr1=np.array([1,2,3,4])
arr3=arr1.view()

print(id(arr1))
print(id(arr3))

print()
print(id(arr1[2]))
print(id(arr3[2]))

print()
print(arr1.ctypes.data)
print(arr3.ctypes.data)
```
